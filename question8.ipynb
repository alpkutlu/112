```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
df = pd.read_csv(r"C:\Users\lasto\PycharmProjects\pythonProject\.venv\Marketing-bank.csv", delimiter=";")
#  Group by 'CampaignType' and sum the spendings
grouped_data = df.groupby('Campaign_Type')['Total_Spending'].sum().reset_index()

# Plot the bar chart
ax = grouped_data.plot(kind='bar', x='Campaign_Type', y='Total_Spending', legend=False)

# Customize the plot
ax.set_title('Total Spendings by Campaign Type')
ax.set_xlabel('Campaign Type')
ax.set_ylabel('Total Spendings')

#Show the plot
plt.show()





```
