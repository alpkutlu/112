```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Read the CSV file with the correct delimiter

```python
df = pd.read_csv('Marketing-bank.csv', delimiter=';')
```

Create a line plot to visualize the relationship between Conversion Rate, Campaign Type, and Target Audience

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x='Campaign_Type', y='Conversion_Rate', hue='Target_Audience', data=df, ci=None, palette='viridis',
             markers=True, dashes=False)
plt.title('Conversion Rate by Campaign Type and Target Audience')
plt.xlabel('Campaign Type')
plt.ylabel('Conversion Rate')
```

Add a legend to differentiate between Target Audiences

```python
plt.legend(title='Target Audience', loc='upper left', bbox_to_anchor=(1, 1))
```

Display the plot

```python
plt.show()
```
