```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Reading the CSV file with the correct delimiter

```python
df = pd.read_csv('Marketing-bank.csv', delimiter=';')
```

Converting the 'Year' column to datetime format

```python
df['Year'] = pd.to_datetime(df['Year'], format='%Y')
```

Calculating the average spend for each year

```python
average_spend_by_year = df.groupby(df['Year'].dt.year)['Total_Spending'].mean().reset_index()
```

Creating a line plot to visualize the average spend for each year

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='Year', y='Total_Spending', data=average_spend_by_year, marker='o', color='b')
plt.title('Average Spending for Each Year')
plt.xlabel('Year')
plt.ylabel('Average Spending')
```

Set the y-axis limits

```python
plt.ylim(2000, 8000)
```

```python
plt.show()
```
