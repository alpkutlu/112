```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Reading the CSV file with the correct delimiter

```python
df = pd.read_csv('Marketing-bank.csv', delimiter=';')
```

Controlling the structure of the dataset

```python
print(df.head())
```

Calculating the average age for each campaign type

```python
average_age_by_campaign = df.groupby('Campaign_Type')['Customer_Age'].mean().reset_index()
```

Sorting the DataFrame by average age in descending order

```python
average_age_by_campaign_sorted = average_age_by_campaign.sort_values(by='Customer_Age', ascending=False)
```

Creating a bar chart to visualize the average age distribution for different types of campaigns

```python
plt.figure(figsize=(10, 6))
bar_plot = sns.barplot(x='Campaign_Type', y='Customer_Age', data=average_age_by_campaign_sorted, palette='viridis')
plt.title('Average Age Distribution Among Different Campaign Types')
plt.xlabel('Campaign Type')
plt.ylabel('Average Age')
```

```python
plt.show()
```
